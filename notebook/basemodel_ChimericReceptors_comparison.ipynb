## Chimeric receptors simulations

Simulations of the base T cell model upon the same initial conditions. The 'idens' reflect the distinct chimeric receptors or lack thereof and T cell identity (CD4 or CD8). Cells simulated here:
* TEGs (gdTCR) 
* NKG2D-41BB co-receptor
* NKG2D-CD28 co-receptor
* CAR
* abTCR
* 103-41BB
* 103-truncated

=> each of them CD4 or CD8

-> final state probabilities for phenotypes

-> final node probabilities for all nodes

```python
import maboss
import pandas as pd
```

Import model

```python
bnd = "./model_Tcell.bnd"
cfg =  "./model_Tcell.cfg"
    
# Load the model
model = maboss.load(bnd, cfg)

on=[0,1]
off=[1,0]
random=[0.5,0.5]
high=[0.2,0.8]


#set parameters
param = {'max_time': 1000, 'time_tick': 1, 'sample_count': 1000, 'thread_count':6, 'use_physrandgen':0}
model.param.update(param)
nodes = list(model.network.keys())
print(nodes)
class inputset:
  def __init__(self, set, value):
    self.set = set
    self.value = value
```

Define outputs and initial conditions to be testes

```python
outputs_1=["cytotoxic", "Apoptosis", "proliferation", "AKT", "IFNg", "NFKB", "PD1", "LAG3", "BLIMP1", "NFAT", "AP1", "ATF2", "TIM3", "Exhaustion","Tem", "Tcm_naive",
           "IL2", "Th1", "Th2", "FOXO1", "NFAT", "TBET", "ERK", "Treg", "kill_potential","AP1", "NFAT"]
outputs_2=["cytotoxic", "Apoptosis", "proliferation", "Th1", "Th2", "Treg", "Tem", "Tcm_naive","Exhaustion", "kill_potential"]
survival=["Apoptosis", "proliferation"]


inputs_environment= ['IFNG_L' ,'IL12_L','IL27_L','IL2_L','IL4_L','TGFB_L', 'CD80_86', 'L41BB', 'FAS_L', 'IL7_L',
                     'TIM3_L','TIGIT_L', 'PD1_L', 'ICOS_L']

# initial conditions for different constructs
iden_TEG_NKG2D_41BB = ['iden_NK41BB', 'iden_gdTCR']
iden_TEG_NKG2D_CD28 = ['iden_NKCD28', 'iden_gdTCR']
iden_TEG_R103_41BB = ['iden_R103', 'iden_gdTCR']
iden_TEG_103_truncated = ['iden_TR103', 'iden_gdTCR']
iden_CAR = ['iden_CAR']
iden_abTCR=['iden_abTCR']
iden_TEG001=['iden_gdTCR']
iden_BTN3A_cis=['iden_BTN3A_cis']
iden_CD4=['iden_CD4']
iden_CD8=['iden_CD8']

CDs=['CD4', 'CD8']
iden_CDs=[iden_CD4, iden_CD8]

idens= iden_TEG_103_truncated + iden_TEG_NKG2D_41BB + iden_TEG_NKG2D_CD28 + iden_TEG_R103_41BB + iden_TEG001 + iden_CAR + iden_abTCR + iden_BTN3A_cis + iden_CD4 + iden_CD8

high=[0.2, 0.8]
low=[0.8, 0.2]
on=[0,1]
off=[1,0]
random=[0.5,0.5]

#ic to simulate T cells with elements of the TME
ic_TME = ['LCK', 'CD45', 'CD8', 
          'TIM3_L', 'IFNG_L','IL2_L','IL4_L','TGFB_L','L41BB', 'FAS_L','TIGIT_L','PD1_L',
          'IL7_L','CD80_86', 'IL12_L','IL27_L','ICOS_L']

#ic to simulate T cells with MM cells
ic_MM = ['LCK', 'CD45', 'CD8', 
         'TIM3_L', 'IFNG_L','IL2_L','IL4_L','TGFB_L','L41BB','FAS_L', 'TIGIT_L', 'PD1_L']

#ic to simulate T cells exposed to antigen
ic_antigen = ['LCK', 'CD45', 'CD8']
```

```python
ident_CART=[inputset(nodes, random), inputset(idens, off),inputset(iden_CAR, on)]
ident_Tcell=[inputset(nodes, random),inputset(idens, off),inputset(iden_abTCR, on)]
ident_TEG001=[inputset(nodes, random),inputset(idens, off),inputset(iden_TEG001, on)]
ident_TEG103=[inputset(nodes, random),inputset(idens, off),inputset(iden_TEG_R103_41BB, on)]
ident_TEG103_trunc=[inputset(nodes, random),inputset(idens, off),inputset(iden_TEG_103_truncated, on)]
ident_NKG2D_CD28=[inputset(nodes, random),inputset(idens, off),inputset(iden_TEG_NKG2D_CD28, on)]
ident_NKG2D_41BB=[inputset(nodes, random),inputset(idens, off),inputset(iden_TEG_NKG2D_41BB, on)]


cell_types=[ident_CART, ident_Tcell, ident_TEG001, ident_TEG103, ident_TEG103_trunc, ident_NKG2D_CD28, ident_NKG2D_41BB]
cell_types_names=['CART', 'Tcell', 'TEG001', 'TEG103', 'TEG103_trunc', 'NKG2D_CD28', 'NKG2D_41BB']

cell_types_dict = dict(zip(cell_types_names, cell_types))
```

```python
#run simulations for all the cell types/constructs under different input_ic

def simulations_NodeStates_prob(input_ic, filename): 

    prob_node = pd.DataFrame()
    prob_state = pd.DataFrame()  
        
    sim = model.copy()
    
    #for cd4 and cd8
    for cd in CDs: 
        # go over different constructs
        for cell in cell_types_names:
            #set ic that define each cell construct
            for input_set in cell_types_dict[cell]:
                maboss.set_nodes_istate(sim, input_set.set , input_set.value)

            if (cd =="CD4"):
                #set ic that define each cell type
                maboss.set_nodes_istate(sim, ['iden_CD4'], on)
            
            if (cd =="CD8"):
                #set ic that define each cell type
                maboss.set_nodes_istate(sim, ['iden_CD8'], on)

            #set ic that define the environment
            maboss.set_nodes_istate(sim, input_ic.set , input_ic.value)


            #set outpus1 -> node
            sim.network.set_output(nodes)
            res = sim.run(cmaboss=True)
            tmp_prob_node=res.get_last_nodes_probtraj()
            tmp_prob_node.index=[cell+'_'+cd]
            prob_node = pd.concat([prob_node, tmp_prob_node], ignore_index = False)
            del(tmp_prob_node)

    
            #set phenotypes -> state
            sim.network.set_output(outputs_2)
            res = sim.run(cmaboss=True)
            tmp_prob_state=res.get_last_states_probtraj()
            tmp_prob_state.index=[cell+'_'+cd]
            prob_state = pd.concat([prob_state, tmp_prob_state], ignore_index = False)
            del(tmp_prob_state)
                
    prob_node.to_csv(filename+"_nodes.csv", index=True) 
    prob_state.to_csv(filename+"_states.csv", index=True) 
```

```python
environment_antigenHigh=inputset(['BCMA', 'pMHCII','pMHCI', 'BTN2_3', 'NKG2D'], high)
environment_antigenRandom=inputset(['BCMA', 'pMHCII','pMHCI', 'BTN2_3', 'NKG2D'], random)

simulations_NodeStates_prob(environment_antigenHigh, 'base_chimeric_receptors_EnvRANDOM_antigenHigh')

simulations_NodeStates_prob(environment_antigenRandom, 'base_chimeric_receptors_EnvRANDOM_antigenRandom')
```
